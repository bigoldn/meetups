```python
from scipy import optimize
```

We are going to solve problems of the following form

$\min_x f(x)$

- Unconstrained optimization
- Multivariate optimization
- Convex/Non-convex optimization

A point $x^* $ is a global minimizer if $f(x^*) \leq f(x)$ for all $x$.

A point is a local minimizer if there is a neighborhood $N$ such that $f(x^*) \leq f(x)$ for all $x \in N$.

In order to check whether a point is a local minimizer we don't need to "check out out the neighborhood" necessarily; if the function is smooth, we can use the gradient and Hessian of $f$ (proof by Taylor expansion of the function). Example function we will be working with:

```python
def rosenbrock(x,y):
     return 100.0 * (y-x**2)**2 + (1-x)**2

x = arange(-2, 2, 0.01)
y = arange(-1, 3, 0.01)
[X, Y] = meshgrid(x, y)
Z = rosenbrock(X, Y)
CS = imshow(Z, interpolation='bilinear', origin='lower', extent=[-2,2,-1,3])
_ = colorbar(CS, shrink=0.8, extend='both')
```

General strategy for solving optimization problems: start at a reasonable point, iteratively find better points. 

__Line Search__

We slice the rosenbrock function at $y = 2$ and find a 1D method.

```python
xs = arange(-2, 2, 0.01)
zs = array([rosenbrock(x, 2.0) for x in xs])
```

```python
x0 = array([-1.7, 2.0])
p0 = array([1.0, 0.0])

def f(x):
    plot(x[0], optimize.rosen(x), 'kx')
    return optimize.rosen(x)

plot(xs, zs)

alpha, _, _, _, _, _ = optimize.line_search(f=f, myfprime=optimize.rosen_der, xk=x0, pk=p0)

plot(-1.5+alpha, optimize.rosen(array([-1.5+alpha,2.0])), 'ro')
```

__Gradient Descent__

Let's do standard gradient descent while using the line search method illustrated above.

```python
x = array([-1.8, 0.9])
p = optimize.rosen_der(x)

imshow(Z, interpolation='bilinear', origin='lower', extent=[-2,2,-1,3])

pn_gd = []
for iter in xrange(5000):
    alpha, _, _, _, _, _ = optimize.line_search(optimize.rosen, optimize.rosen_der, xk=x, pk=p)
    x -= alpha * p
    p = optimize.rosen_der(x)
    if iter % 100 == 0:
        plot(x[0], x[1], 'k.')
    
    pn_gd.append(np.dot(p,p))
    
    if np.dot(p,p) < 1e-14:
        break
print "Number of iterations = %d." % (iter+1)
```

We can also look at the squared norm of the gradient (note the log scale).

```python
semilogy(pn_gd)
```

Rather than using only first order information to determine the direction of steepest descent, we can also use second order information to understand rates of change of the direction of steepest descent. We call this the Newton iteration.

```python
x = array([-1.8, 0.9])
p = optimize.rosen_der(x)
H = optimize.rosen_hess(x)

imshow(Z, interpolation='bilinear', origin='lower', extent=[-2,2,-1,3])

pn_n = []
for iter in xrange(1000):
    x -= linalg.solve(H, p)
    p = optimize.rosen_der(x)
    H = optimize.rosen_hess(x)
    plot(x[0], x[1], 'k.')
    
    pn_n.append(np.dot(p,p))
    
    if np.dot(p,p) < 1e-14:
        break
print "Number of iterations = %d." % (iter+1)
```

```python
semilogy(pn_n)
```

Sometimes it is hard to compute the Hessian, or the dimensionality of the optimization is so large that it is impossible to compute or store the Hessian. In this case, we can try to approximate the Hessian using a low rank matrix.

__BFGS & L-BFGS__

The BFGS (or Broyden-Fletcher-Goldfarb-Shanno) algorithm is an algorithm when the Hessian is hard to compute, but it is still possible to store  a full approximation to the Hessian.

```python
x0 = array([-1.8, 0.9])

def f(x):
    plot(x[0],  x[1], 'k.')
    return optimize.rosen(x)

imshow(Z, interpolation='bilinear', origin='lower', extent=[-2,2,-1,3])

res = optimize.minimize(f, x0, method="BFGS", jac=optimize.rosen_der, options={"disp": True})
```

There is an algorithm called L-BFGS (or Limited Memory BFGS) which is similar in spirit to BFGS but only keeps track of a low rank approximation to the Hessian approximation.

__Conjugate Gradient__

The conjugate gradient algorithm was originally developed to solve linear systems of equations. In the linear case, the algorithm generates a sequence of vectors which are _conjugate_ w.r.t. the matrix we are trying to solve for. The conjugate gradient algorithm can easily be adapted to work for nonlinear functions and has proven to be quite effective in practice.

```python
x0 = array([-1.8, 0.9])

def f(x):
    plot(x[0],  x[1], 'k.')
    return optimize.rosen(x)

imshow(Z, interpolation='bilinear', origin='lower', extent=[-2,2,-1,3])

res = optimize.minimize(f, x0, method="CG", jac=optimize.rosen_der, options={"disp": True})
```
